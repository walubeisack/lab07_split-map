```python
import leafmap
import rasterio
import folium
import leafmap.foliumap as leafmap_folium
```

```python
m = leafmap_folium.Map(center=[40.76, -98.39], zoom=4)

url1 = "https://www.mrlc.gov/geoserver/mrlc_display/nlcd_tcc_conus_2021_v2021-4/wms?"
url2 = "https://www.mrlc.gov/geoserver/mrlc_display/nlcd_tcc_conus_2011_v2021-4/wms?"

left_layer = folium.WmsTileLayer(
    url=url1,
    layers="nlcd_tcc_conus_2021_v2021-4",
    name="NLCD 2021",
    attr="MRLC",
    fmt="image/png",
    transparent=True,
)
right_layer = folium.WmsTileLayer(
    url=url2,
    layers="nlcd_tcc_conus_2011_v2021-4",
    name="NLCD 2011",
    attr="MRLC",
    fmt="image/png",
    transparent=True,
)

left_layer
m.split_map(left_layer, right_layer)
m
```

```python
m.to_html('lab07split_map.html', title = "2021v2011_Tree_canopy")
```
